```python
import pandas as pd
import plotly.express as p


```

```python
x1=pd.read_excel('customer intelliegnce ecommerce edit 2.xlsx')

x1
```

```python
x1.describe()
```

```python

luv=x1.groupby('Brand of the product')['Rating of the product'].mean().reset_index()

fig=p.line(luv,x='Brand of the product',y='Rating of the product')

fig.update_layout(xaxis_title='Brand',yaxis_title='rating of the product',height=600,width=2200,xaxis_tickangle=90)
fig.show()
```

```python

luv2=x1.groupby(['Season','Brand of the product'])['Price of the product'].max().reset_index()
luv2=luv2.sort_values('Price of the product',ascending=False)

fig=p.bar(luv2,x='Brand of the product',y='Price of the product',color='Season')

fig.update_layout(xaxis_title='seasons',yaxis_title='maximum price of the product',height=700,width=2000,xaxis_tickangle=90)

fig.show()
```

```python

luv3=x1.groupby(['Brand of the product','Gender'])['Number of clicks on similar products'].min().reset_index()
luv3=luv3.sort_values('Number of clicks on similar products',ascending=False)
fig=p.bar(luv3,x='Brand of the product',y=('Number of clicks on similar products'),color='Gender')

fig.update_layout(xaxis_title='Brand ',yaxis_title='Minimum number f clicks on similar products',height=700,width=2700,xaxis_tickangle=90)

fig.show()
```

```python
luv4=x1.groupby('Brand of the product')['Price of the product'].min().reset_index()
fig=p.line(luv4,x='Brand of the product',y='Price of the product')

fig.update_layout(xaxis_title='Brand of the product',yaxis_title='Minimum purchasing price (in rupees)',xaxis_tickangle=90,height=600,width=2500)

fig.show()
```

```python
import plotly.express as px
from ipywidgets import interact

def brand_slicer(brand):
    if brand == "All":
        df =x1
    else:
        df = x1[x1["Brand of the product"] == brand]

        fig=px.box(df,x='Brand of the product',y='Rating of the product')
        fig.update_layout(xaxis_title='brand of product',yaxis_title='rating of the product',xaxis_tickangle=90,height=2000)
        fig.show()


   
interact(
    brand_slicer,
    brand=["All"] + sorted(x1["Brand of the product"].dropna().unique().tolist())
)










```

```python
import plotly.express as px
from ipywidgets import interact


l5=x1.groupby(['brand of the product','customer segment'])['number of similar products purchased so far'].min().reset_index()

def brand_slicer(brand):
    if brand == "All":
        df = l5
    else:
        df = l5[l5["Brand of the product"] == brand]

    fig = px.bar(
        df,
        x="brand of the product",
        y="number of similar products purchased so far",
        color="customer segment"
    )
    fig.update_layout(height=500,width=3000)
    fig.show()

interact(
    brand_slicer,
    brand=["All"] + sorted(x1["brand of the product"].dropna().unique().tolist())
)
```
